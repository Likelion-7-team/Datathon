```python
new_ids = set(new['SK_ID_CURR'])
```

```python
credit_score['group'] = credit_score['SK_ID_CURR'].apply(
    lambda x: '신규' if x in new_ids else '기존'
)
```

```python
avg_scores = credit_score.groupby('group')['Scaled_Score'].mean()
print(avg_scores)
```

```python
# 1. 수치형 컬럼만 추출
num_cols = credit_score.select_dtypes(include='number').columns.drop('SK_ID_CURR')

# 2. 그룹별 평균 계산
grouped_means = credit_score.groupby('group')[num_cols].mean()

# 3. 평균 차이 계산 (신규 - 기존)
mean_diff = grouped_means.loc['신규'] - grouped_means.loc['기존']

# 결과 보기
print(mean_diff.sort_values(ascending=False))  # 차이 큰 순으로 보기
```
