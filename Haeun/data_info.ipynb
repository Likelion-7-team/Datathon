```python
import pandas as pd
from sqlalchemy import create_engine

# 1. MySQL 연결 설정
user = "root"
password = "root"
host = "localhost"
port = 3306
database = "finance"

engine = create_engine(f"mysql+pymysql://{user}:{password}@{host}:{port}/{database}")

# 2. CSV 파일들 로드
df1 = pd.read_csv('./data/application_test.csv', encoding='utf-8-sig')
df2 = pd.read_csv('./data/application_train.csv', encoding='utf-8-sig')
df3 = pd.read_csv('./data/bureau.csv', encoding='utf-8-sig')
df4 = pd.read_csv('./data/bureau_balance.csv', encoding='utf-8-sig')
df5 = pd.read_csv('./data/credit_card_balance.csv', encoding='utf-8-sig')
df6 = pd.read_csv('./data/installments_payments.csv', encoding='utf-8-sig')
df7 = pd.read_csv('./data/POS_CASH_balance.csv', encoding='utf-8-sig')
df8 = pd.read_csv('./data/previous_application.csv', encoding='utf-8-sig')

# 3. 테이블로 삽입
df1.to_sql("application_test", engine, if_exists='replace', index=False)
df2.to_sql("application_train", engine, if_exists='replace', index=False)
df3.to_sql("bureau", engine, if_exists='replace', index=False)
df4.to_sql("bureau_balance", engine, if_exists='replace', index=False)
df5.to_sql("credit_card_balance", engine, if_exists='replace', index=False)
df6.to_sql("installments_payments", engine, if_exists='replace', index=False)
df7.to_sql("POS_CASH_balance", engine, if_exists='replace', index=False)
df8.to_sql("previous_application", engine, if_exists='replace', index=False)

print("모든 CSV 파일이 finance 데이터베이스에 성공적으로 삽입되었습니다.")
```
